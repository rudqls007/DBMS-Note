## Fancy indexing
- 데이터를 filtering
- 전체데이터에서 원하는 일부의 데이터를 찾아오는 법

```python
#df["X1"]
df.X1
```

```python
# df에서 index value를 기준으로 indexing도 가능하다.
#df. index[0]
df.loc[0]
```

```python
df
```

```python
# df loc는 특정값을 기준으로 indexing한다.
df.loc[7]
```

```python
#df.loc에 2차원 indexing도 가능하다.
df.loc[9,'X4']
```

```python
df.loc[3:5, 'X2':'X4']
```

```python
df.loc[[0, 1, 4, 6, 10]][['X1', 'X3']]
```

```python
# 2차원 리스트 indexing 과 같은 원리
# indexing location based indexing
df.iloc[5]
```

```python
# iloc로 2차원 indexing
# row 기준으로 index 3, 4를 가져오고,
# column 기준으로 0,1을 가져온다.
df.iloc[[3,4],[0,1]]
```
